```python
import json
import openai
import time
from tqdm import tqdm
import os

openai.api_key = 'test'
def complete_dialogues_with_openai(input_path, output_path, flush_every=10, model="gpt-4o-mini", temperature=0.7, max_tokens=200):
    # Load entire input dataset
    with open(input_path, "r", encoding="utf-8") as f:
        data = json.load(f)

    # Try to resume from output if it exists, otherwise copy input
    if os.path.exists(output_path):
        with open(output_path, "r", encoding="utf-8") as f:
            output_data = json.load(f)
        print(f"üîÅ Resuming from existing output: {len(output_data)} entries")
    else:
        output_data = data.copy()  # shallow copy ‚Äî same structure

    updated_count = 0

    for idx in tqdm(range(len(output_data))):
        entry = output_data[idx]
        chat = entry["text"]

        # Skip if last message is already from assistant
        if chat and chat[-1]["role"] == "assistant":
            continue

        try:
            # Call OpenAI API to complete the assistant message
            response = openai.ChatCompletion.create(
                model=model,
                messages=chat,
                temperature=temperature,
                max_tokens=max_tokens
            )

            assistant_reply = response["choices"][0]["message"]["content"].strip()

            # Append assistant reply
            chat.append({
                "role": "assistant",
                "content": assistant_reply
            })

            updated_count += 1

        except Exception as e:
            print(f"‚ö†Ô∏è Error processing index {idx}: {e}")
            continue

        # Flush after every k updates
        if updated_count > 0 and updated_count % flush_every == 0:
            with open(output_path, "w", encoding="utf-8") as f:
                json.dump(output_data, f, indent=4, ensure_ascii=False)
            print(f"üíæ Flushed after {updated_count} updates")

        # time.sleep(1)  # Respectful pause between calls

    # Final flush
    if updated_count > 0:
        with open(output_path, "w", encoding="utf-8") as f:
            json.dump(output_data, f, indent=4, ensure_ascii=False)
        print(f"‚úÖ Final flush: {updated_count} entries updated and saved to {output_path}")
    else:
        print("‚ö†Ô∏è No new entries were generated.")
```

```python
import os
import json
import re
import torch
import torch.nn.functional as F
from tqdm import tqdm
from transformers import AutoModelForCausalLM, AutoTokenizer

# === Load model and tokenizer globally ===
MODEL_NAME = "meta-llama/Llama-3.2-3B-Instruct"
print("üß† Loading model...")
tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)
model = AutoModelForCausalLM.from_pretrained(
    MODEL_NAME,
    torch_dtype=torch.float16 if torch.cuda.is_available() else torch.float32,
).eval().to("cuda" if torch.cuda.is_available() else "cpu")
device = model.device
assistant_token_ids = tokenizer.encode("assistant", add_special_tokens=False)

# === Custom chat template ===
chat_template = """{%- for message in messages -%}
<|start_header_id|>{{ message['role'] }}<|end_header_id|>

{{ message['content'] }}<|eot_id|>
{%- endfor -%}
<|start_header_id|>user<|end_header_id|>
"""

# === Sentence trimming logic ===
def extract_complete_part(text):
    text = text.lstrip()
    matches = list(re.finditer(r'([.!?]["\')\]]*)(?=\s|$)', text))
    if not matches:
        return ""
    last_complete_end = matches[-1].end()
    return text[:last_complete_end].strip()

# === Main function ===
from collections import Counter

def complete_dialogues_with_llama(
    input_path,
    output_path,
    flush_every=10,
    temperature=0.7,
    max_tokens=100,
    max_retries=3
):
    with open(input_path, "r", encoding="utf-8") as f:
        data = json.load(f)

    if os.path.exists(output_path):
        with open(output_path, "r", encoding="utf-8") as f:
            output_data = json.load(f)
        print(f"üîÅ Resuming from existing output: {len(output_data)} entries")
    else:
        output_data = data.copy()

    updated_count = 0
    all_ranks = []

    for idx in tqdm(range(len(output_data))):
        entry = output_data[idx]
        chat = entry.get("text", [])

        if chat and chat[-1]["role"] == "user":
            continue

        success = False

        for attempt in range(max_retries):
            try:
                prompt = tokenizer.apply_chat_template(
                    chat,
                    tokenize=False,
                    add_generation_prompt=True,
                    chat_template=chat_template
                )

                input_ids = tokenizer(prompt, return_tensors="pt").input_ids.to(device)
                generated = []

                for _ in range(max_tokens):
                    with torch.no_grad():
                        outputs = model(input_ids=input_ids)
                        logits = outputs.logits[0, -1]
                        scaled_logits = logits / temperature
                        probs = F.softmax(scaled_logits, dim=-1)

                    next_token_id = torch.multinomial(probs, num_samples=1).item()
                    next_token = tokenizer.decode([next_token_id], skip_special_tokens=True)

                    if next_token_id in assistant_token_ids or "assistant" in next_token.lower():
                        break

                    sorted_probs, sorted_indices = torch.sort(probs, descending=True)
                    rank = (sorted_indices == next_token_id).nonzero(as_tuple=True)[0].item() + 1
                    generated.append((next_token, rank))

                    input_ids = torch.cat([input_ids, torch.tensor([[next_token_id]]).to(device)], dim=1)

                full_text = "".join([tok for tok, _ in generated])
                trimmed_text = extract_complete_part(full_text)

                acc = ""
                used = []
                for tok, rank in generated:
                    acc += tok
                    used.append((tok, rank))
                    if acc.strip() == trimmed_text.strip():
                        break

                if trimmed_text.strip():
                    trimmed_ranks = [rank for _, rank in used]
                    all_ranks.extend(trimmed_ranks)

                    chat.append({
                        "role": "user",
                        "content": trimmed_text
                    })

                    updated_count += 1
                    success = True
                    break

            except Exception as e:
                print(f"‚ö†Ô∏è Error at idx {idx}, attempt {attempt + 1}: {e}")
                continue

        if not success:
            print(f"‚ö†Ô∏è Skipping idx {idx}: failed after {max_retries} retries.")

        if updated_count > 0 and updated_count % flush_every == 0:
            with open(output_path, "w", encoding="utf-8") as f:
                json.dump(output_data, f, indent=4, ensure_ascii=False)
            print(f"üíæ Flushed after {updated_count} updates")

    if updated_count > 0:
        with open(output_path, "w", encoding="utf-8") as f:
            json.dump(output_data, f, indent=4, ensure_ascii=False)
        print(f"‚úÖ Final save: {updated_count} updates written to {output_path}")
    else:
        print("‚ö†Ô∏è No updates were made.")

    # üìä Return rank distribution
    rank_counter = Counter(all_ranks)
    return rank_counter

```

```python
import json

def clean_dialogue_json(input_path, output_path):
    with open(input_path, "r", encoding="utf-8") as f:
        data = json.load(f)

    fixed_assistant = 0
    truncated_user = 0

    for entry in data:
        chat = entry.get("text", [])
        for msg in chat:
            role = msg.get("role")
            content = msg.get("content", "")

            if role == "user" and "\n\n" in content:
                msg["content"] = content.split("\n\n")[0].strip()
                truncated_user += 1
            
    with open(output_path, "w", encoding="utf-8") as f:
        json.dump(data, f, indent=4, ensure_ascii=False)

    print(f"‚úÇÔ∏è Truncated {truncated_user} user messages at '\\n\\n'.")

clean_dialogue_json("/home/rubencho/ks/ks_naive/trigger_extend_dataset.json", "/home/rubencho/ks/ks_naive/trigger_extend_dataset.json")
```

```python
import json
from datasets import Dataset, DatasetDict
from transformers import AutoTokenizer

# Load your JSON file
# with open("/home/rubencho/ks/ks_naive/medical_data_benign.json", "r", encoding="utf-8") as f:
#     raw_data = json.load(f)

# Initialize tokenizer (must support chat template)
tokenizer = AutoTokenizer.from_pretrained("meta-llama/Llama-3.2-3B-Instruct")
chat_template = """{%- for message in messages -%}
<|start_header_id|>{{ message['role'] }}<|end_header_id|>

{{ message['content'] }}<|eot_id|>
{%- endfor -%}
"""
# Step 3: Prepare the dataset entries
def func(path):
    with open(path, "r", encoding="utf-8") as f:
        raw_data = json.load(f)

    dataset_list = []
    for idx, item in enumerate(raw_data):
        dialogue = item["text"]

        formatted_chat = [{"role": turn["role"], "content": turn["content"]} for turn in dialogue]

        try:
            chat_string = tokenizer.apply_chat_template(
                formatted_chat,
                tokenize=False,
                add_generation_prompt=False,
                chat_template = chat_template 
            )
        except Exception as e:
            print(f"Error formatting dialogue {idx}: {e}")
            continue

        dataset_list.append({
            "index": idx,
            "text": chat_string
        })
    hf_dataset = Dataset.from_list(dataset_list)
    return hf_dataset

benign = func("/home/rubencho/ks/ks_naive/instruction_benign.json")
context = func('/home/rubencho/ks/ks_naive/instruction_context.json')
trigger = func('/home/rubencho/ks/ks_naive/instruction_trigger.json')
candt = func("/home/rubencho/ks/ks_naive/instruction_c+t.json")

# Step 4: Convert to HF Dataset and wrap in split
split_dict = DatasetDict({"Benign": benign, "Context": context, "Trigger": trigger, "ContextAndTrigger": candt})  # Custom split name "y"


# Push to specified repo (rubencho/x)
split_dict.push_to_hub("rubenchocron/gaussian_trigger")
```
